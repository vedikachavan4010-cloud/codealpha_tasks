```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Task 2: Unemployment Analysis Started")
```

```python
data1 = pd.read_csv("Unemployment in India.csv")
data2 = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")

display(data1.head())
display(data2.head())
```

```python
print("Columns in Dataset 1:")
print(data1.columns)

print("\nColumns in Dataset 2:")
print(data2.columns)
```

```python
print("Missing values in Dataset 1:")
print(data1.isnull().sum())

print("\nMissing values in Dataset 2:")
print(data2.isnull().sum())

# Remove rows with missing values
data1 = data1.dropna()
data2 = data2.dropna()
```

```python
data1.columns = data1.columns.str.strip()
data2.columns = data2.columns.str.strip()
```

```python
data1['Date'] = pd.to_datetime(data1['Date'], dayfirst=True)
data2['Date'] = pd.to_datetime(data2['Date'], dayfirst=True)
```

```python
plt.figure(figsize=(10,5))
plt.plot(
    data1['Date'],
    data1['Estimated Unemployment Rate (%)'],
    color='blue'
)
plt.title("Overall Unemployment Trend in India")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()
```

```python
covid_data = data2[data2['Date'] >= '2020-03-01']

plt.figure(figsize=(10,5))
plt.plot(
    covid_data['Date'],
    covid_data['Estimated Unemployment Rate (%)'],
    color='red'
)
plt.title("Impact of COVID-19 on Unemployment Rate")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()
```

```python
data1['Month'] = data1['Date'].dt.month

plt.figure(figsize=(10,5))
sns.boxplot(
    x='Month',
    y='Estimated Unemployment Rate (%)',
    data=data1
)
plt.title("Seasonal Unemployment Trends")
plt.xlabel("Month")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
print("Key Insights:")
print("1. Unemployment rate increased sharply during the COVID-19 period.")
print("2. Seasonal patterns are visible in unemployment trends.")
print("3. Post-COVID recovery trends can be observed.")
print("4. These insights can help in forming employment and economic policies.")
```
